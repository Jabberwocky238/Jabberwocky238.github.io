```python
from datetime import datetime
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
from sqlalchemy.ext.declarative import declarative_base

engine = create_engine('sqlite:///endb.db')
Session = sessionmaker(bind=engine)
Base = declarative_base()

from sqlalchemy import DateTime, Column, Integer, String, ForeignKey, Float, Boolean, Text
class English(Base):
    __tablename__ = 'english'

    id = Column(Integer, primary_key=True, autoincrement=True)
    create_date = Column(DateTime, nullable=False)
    word = Column(String, nullable=False)
    translate = Column(String, nullable=False)
    en_sentense = Column(String, nullable=False)

    def __init__(self, word, translate, en_sentense):
        self.create_date = datetime.now()
        self.word = word
        self.translate = translate
        self.en_sentense = en_sentense

    def __repr__(self):
        return f"<English word='{self.word}', translate='{self.translate}'>"
    
# Base.metadata.drop_all(engine) 
# Base.metadata.create_all(engine)
```
